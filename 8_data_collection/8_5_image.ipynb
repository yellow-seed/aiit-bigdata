### 部分の切り抜き

座標 (x, y)
- 左上が原点 (0,0)
- 横方向が x 軸
  - 右方向に増加
- 縦方向が y 軸
  - 下方向に増加

```python
from PIL import Image
import matplotlib.pyplot as plt

im = Image.open('data/2022PBLpamphlet_ura.jpeg')

# 基本情報の確認
w, h = im.size
print('width={}, height={}'.format(w, h))
```

```python
# 座標：左上(230, 1920) 右下(880, 2240)
im1 = im.crop((230, 1920, 880, 2240))

# 基本情報の確認
w, h = im1.size
print('width={}, height={}'.format(w, h))

# 軸等を非表示
plt.axis("off")
# 表示
plt.imshow(im1)
```

### サイズ変更

```python
# 引数はint型
im1_sm = im1.resize((int(w/4), int(h/4)))

# 基本情報の確認
# - size はタプルのため * を付ける
print('width={}, height={}'.format(*im1_sm.size))

# 軸等を非表示
plt.axis("off")
# 表示
plt.imshow(im1_sm)
```

```python
# LANCZOS: 計算は遅いが品質が良い
im1_sm = im1.resize((int(w/4), int(h/4)), Image.Resampling.LANCZOS)

# 軸等を非表示
plt.axis("off")
# 表示
plt.imshow(im1_sm)
```

```python
# NEAREST: 計算は速いが品質は悪い
im1_sm = im1.resize((int(w/4), int(h/4)), Image.Resampling.NEAREST)

# 軸等を非表示
plt.axis("off")
# 表示
plt.imshow(im1_sm)
```

フィルタの種類<br>
https://pillow.readthedocs.io/en/stable/handbook/concepts.html#concept-filters
- デフォルト：PIL.Image.Resampling.BICUBIC

### 回転

```python
# expand: True を指定することで元画像のサイズのまま回転
im1_rot = im1.rotate(90, expand=True)

# 基本情報の確認
print('width={}, height={}'.format(*im1_rot.size))

# 軸等を非表示
plt.axis("off")
# 表示
plt.imshow(im1_rot)
```

```python
# expand: 指定しないと元画像のサイズを維持
im1_rot = im1.rotate(90)

# 基本情報の確認
print('width={}, height={}'.format(*im1_rot.size))

# 軸等を非表示
plt.axis("off")
# 表示
plt.imshow(im1_rot)
```

### グレースケール変換、モノクローム変換

```python
# L: グレースケール
im1_gray = im1.convert('L')

# グレースケールで表示
plt.gray()
# 軸等を非表示
plt.axis("off")
# 表示
plt.imshow(im1_gray)
```

```python
# 1: モノクロ
im1_bw = im1.convert('1')

# 軸等を非表示
plt.axis("off")
# 表示
plt.imshow(im1_bw)
```
