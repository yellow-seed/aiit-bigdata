2. メソッドによるテキストの取得

````python
from pdfminer.high_level import extract_text

text = extract_text('data/本学学生の皆さんへ.pdf')
# 確認
text
````

3. 行ごとの分割、不要な改行コードの削除

````python
text2 = []

for l in text.split('\n\n'):
    text2.append(l.replace('\n', ''))

text2
````
